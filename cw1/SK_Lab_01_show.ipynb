**Zadanie 4.** W jednym oknie wygeneruj cztery wykresy 2D/3D funkcji $f(x,y) = \cos(x)|\sin(y)|$ dla $(x, y) \in [-2\pi, 2\pi]^2$ według podanych wymagań:
- wykres siatki 3D
- wykres powierzchni 3D
- wykres rzutu funkcji na płaszczyznę z konturami (sprawdź parametr *levels*)
- wykres powierzchni 3D z izoliniami

```python
x = np.linspace(-2*np.pi, 2*np.pi,100)
y = np.linspace(-2*np.pi, 2*np.pi,100)

X, Y = np.meshgrid(x, y)
Z = np.cos(X) * np.abs(np.sin(Y))

fig = plt.figure(figsize = (12,12))

ax1 = fig.add_subplot(2,2,1, projection="3d")
ax1.plot_wireframe(X,Y,Z)
ax1.set_title("wykres siatki 3D")

ax2 = fig.add_subplot(2,2,2, projection="3d")
surf = ax2.plot_surface(X,Y,Z, cmap=cm.viridis)
fig.colorbar(surf, ax=ax2)
ax2.set_title("wykres powierzchni 3D")

ax3 = fig.add_subplot(2,2,3)
contour = ax3.contourf(X,Y,Z, cmap=cm.viridis)
fig.colorbar(contour, ax=ax3)
ax3.set_title("wykres rzutu funkcji na plaszczyznie z konturami")

ax4 = fig.add_subplot(2,2,4, projection="3d")
ax4.plot_wireframe(X,Y,Z)
ax4.contour(X,Y,Z, cmap = cm.viridis, offset=np.min(Z))
ax4.set_title("wykres powierzchni 3d z izoliniami")
```

**Zadanie 5.** Narysuj krzywą daną parametrycznie:
- $x = \tan \frac{t}{2}$
- $y = \cos \frac{t}{2}$
- $z = \frac{t}{5}$

dla $t\in [0,10\pi]$

```python
t = np.linspace(0,10*np.pi,1000)

x = np.tan(t/2)
y = np.cos(t/2)
z = t/5

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection = "3d")
ax.plot(x,y,z)

ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")
ax.zaxis.labelpad= -3

ax.set_title("wykres krzywej parametrycznej")
plt.show()
```

**Zadanie 6.** Utworzyć wykres powierzchni $z=\cos{\left(x^\frac{1}{y}\right)}+\sin{\left(y^\frac{1}{x}\right)}$ dla dodatnich $x$ i $y$, takich, że $x,y\leq 5$. Wykreślić izolinie na tym wykresie i ustawić kamerę na najwyższy punkt wykresu.

```python
x = np.linspace(0.1,5,1000)
y = np.linspace(0.1,5,1000)

X,Y = np.meshgrid(x,y)
Z = np.cos(X**(1/y)) + np.sin(Y**(1/X))

fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot(111, projection = "3d")

surf = ax.plot_surface(X,Y,Z, cmap = cm.viridis, alpha=0.5)
ax.contour(X,Y,Z, cmap = cm.coolwarm, offset = np.min(Z))
ax.set_title("powierzchnia z izoliniami")
ax.view_init(elev=45, azim=135)
plt.show()
```

**Zadanie 7.** Utwórz wykres z zadania poprzedniego, tak aby był w odcieniach szarości. Izolinie podostaw w kolorze.

**Zadanie 9.** Na podstawie zadania 2 utwórz animację rysowania wykresów trzech funkcji i zapisz jako .gif.

```python
interval = 100
frames = 100

x = np.linspace(0, 2 * np.pi)
f = np.sin(x)
g = np.sin(2 * x)
h = np.sin(3 * x )

fig, ax = plt.subplots()
ax.set_xlim(0,2*np.pi)
ax.set_ylim(-1,1)

(line_f, ) = ax.plot([],[],"orange", linestyle = ":", label = "funkcja $f(x)$")
(line_g, ) = ax.plot([],[],"green", label = "funkcja $g(x)$")
(line_h, ) = ax.plot([],[],"red", linestyle = "--", label = "funkcja $h(x)$")


ax.set_title("wykres trzech funkcji")

ax.set_xlabel("Oś $x$")
ax.set_ylabel("Oś $y$")

ax.grid()

def update(frame):
    line_f.set_data(x[:frame+1],f[:frame+1])
    line_g.set_data(x[:frame+1],g[:frame+1])
    line_h.set_data(x[:frame+1],h[:frame+1])

anim = FuncAnimation(fig, update, frames=frames, interval=interval)
anim.save("zad_9.gif")
```
